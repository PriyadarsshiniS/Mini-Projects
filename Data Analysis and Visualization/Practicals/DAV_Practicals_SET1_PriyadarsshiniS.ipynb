### INFERENCE: <font color='blue'> In the given Heart dataset,there are totally 7 columns with outliers. And the columns  are 'trestbps', 'chol', 'fbs', 'thalach','oldpeak','ca', 'thal'.</font>

```python
print('Before dropping outlier=',data.shape)

#Removing outliers column wise
#trestbps
def find_outliers_IQR(df1):
    q1=df1.quantile(0.25)
    q3=df1.quantile(0.75)
    IQR=q3-q1
    outliers=[x for x in df1 if (x <(q1-(1.5*IQR))) | (x>(q3+(1.5*IQR)))]
    return outliers
outliers = find_outliers_IQR(data["trestbps"])
outliers_index=np.where(data["trestbps"].isin(outliers))
outliers_index=list(outliers_index[0])
data.drop(index=outliers_index, inplace=True)
data.reset_index(drop=True,inplace=True)


#chol
def find_outliers_IQR(df1):
    q1=df1.quantile(0.25)
    q3=df1.quantile(0.75)
    IQR=q3-q1
    outliers=[x for x in df1 if (x <(q1-(1.5*IQR))) | (x>(q3+(1.5*IQR)))]
    return outliers
outliers = find_outliers_IQR(data["chol"])
outliers_index=np.where(data["chol"].isin(outliers))
outliers_index=list(outliers_index[0])
data.drop(index=outliers_index, inplace=True)
data.reset_index(drop=True,inplace=True)

#fbs
def find_outliers_IQR(df1):
    q1=df1.quantile(0.25)
    q3=df1.quantile(0.75)
    IQR=q3-q1
    outliers=[x for x in df1 if (x <(q1-(1.5*IQR))) | (x>(q3+(1.5*IQR)))]
    return outliers
outliers = find_outliers_IQR(data["fbs"])
outliers_index=np.where(data["fbs"].isin(outliers))
outliers_index=list(outliers_index[0])
data.drop(index=outliers_index, inplace=True)
data.reset_index(drop=True,inplace=True)

#oldpeak
def find_outliers_IQR(df1):
    q1=df1.quantile(0.25)
    q3=df1.quantile(0.75)
    IQR=q3-q1
    outliers=[x for x in df1 if (x <(q1-(1.5*IQR))) | (x>(q3+(1.5*IQR)))]
    return outliers
outliers = find_outliers_IQR(data["oldpeak"])
outliers_index=np.where(data["oldpeak"].isin(outliers))
outliers_index=list(outliers_index[0])
data.drop(index=outliers_index, inplace=True)
data.reset_index(drop=True,inplace=True)


#thalach
def find_outliers_IQR(df1):
    q1=df1.quantile(0.25)
    q3=df1.quantile(0.75)
    IQR=q3-q1
    outliers=[x for x in df1 if (x <(q1-(1.5*IQR))) | (x>(q3+(1.5*IQR)))]
    return outliers
outliers = find_outliers_IQR(data["thalach"])
outliers_index=np.where(data["thalach"].isin(outliers))
outliers_index=list(outliers_index[0])
data.drop(index=outliers_index, inplace=True)
data.reset_index(drop=True,inplace=True)


#ca
def find_outliers_IQR(df1):
    q1=df1.quantile(0.25)
    q3=df1.quantile(0.75)
    IQR=q3-q1
    outliers=[x for x in df1 if (x <(q1-(1.5*IQR))) | (x>(q3+(1.5*IQR)))]
    return outliers
outliers = find_outliers_IQR(data["ca"])
outliers_index=np.where(data["ca"].isin(outliers))
outliers_index=list(outliers_index[0])
data.drop(index=outliers_index, inplace=True)
data.reset_index(drop=True,inplace=True)


#thal
def find_outliers_IQR(df1):
    q1=df1.quantile(0.25)
    q3=df1.quantile(0.75)
    IQR=q3-q1
    outliers=[x for x in df1 if (x <(q1-(1.5*IQR))) | (x>(q3+(1.5*IQR)))]
    return outliers
outliers = find_outliers_IQR(data["thal"])
outliers_index=np.where(data["thal"].isin(outliers))
outliers_index=list(outliers_index[0])
data.drop(index=outliers_index, inplace=True)
data.reset_index(drop=True,inplace=True)

print('After dropping outlier=',data.shape)
```
